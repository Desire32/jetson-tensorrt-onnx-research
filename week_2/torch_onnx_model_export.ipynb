```python
import torch.onnx
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-4.0-1b")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-4.0-1b")
```

```python
model.eval();
```

```python
input_tensor = tokenizer("Hello, my name is", return_tensors="pt")["input_ids"]
```

```python
torch.onnx.export(
    model,
    (input_tensor,),
    "./granite-1b/granite-4.0-h-1b.onnx",
    input_names=["input"],
    output_names=["output"],
    opset_version=17,
    external_data=True
)
```
